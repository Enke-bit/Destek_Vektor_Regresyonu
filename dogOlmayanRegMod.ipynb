```python
# Gerekli olan kütüpanelerin import edilmesi.

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
# Sklearn kütüpanesinden gerekli model import edilmesi işlemi yapıldı.
from sklearn.svm import SVR
```

```python
# Bazı algoritmaları çalıştırırken hata olarak görünüyor fakat çalışıyor bu sorunun önüne geçemk için aşağıdaki işlemi gerçekleştiriyorum.

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Veri setinin okunmasını sağlıyorum.
df = pd.read_csv(r"C:\Users\realb\Desktop\ML_dersler\ML_301\Destek Vektör Regresyonu\Hitters.csv")

# Veri setinin içinden eksik gözlemleri çıkarmak gerekiyor bu işlemi yapıyorum.
df = df.dropna()

# Şimdi ise dummies ile kukla veri işlemi yapıyorum. One end coding yöntemi ile bu işlemi yapıyorum.
dms = pd.get_dummies(df[['League', 'Division','NewLeague']])

# Bağımlı değişkenimi "y" atıyorum.
y = df["Salary"]

# Veri setimiz içindeki bağımlı değişkeni ve katogorik değişkenlerin götürülmesi işlemini yapıyorum.
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')

# Sonuçunda yukardaki değerlerin dummi versyonlarını ve bağımsız değişkenlerimi bir arada tutma işlemini yapıyorum.
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

# Şimdi ise veri setimizde train ve test ayrımı yapmak için bu ayrımı yapma işlemini yapıyorum.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Burda veri setinde eğer verilerin doğrusal olmadığını düşünürsek burda linear ifadesini değiştirerek başka bir şekilde kulanabiliriz.
svr_model = SVR(kernel="linear").fit(X_train, y_train)
```

```python
# X test veri gurubundaki bağımsız değişkenlerden tahmin yapıp bunun RMSE değerini hesaplamasını yapıyorum.
y_pred = svr_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

Model Tuning

```python
svr_model = SVR(kernel = "linear")
```

```python
help(svr_model)
```

```python
# Şimdi ben burda yapmak istediğim şu ben burada olası C değerlerini vererek ve bunları k katlı çağraz doğrulama ile işleme alarak olası daha iyi sonuçları ele alıyorum.

# Burda ben C ceza değerlerini paremetre olarak gireçeğim ve bunu olası değerler olarak girilmesini sağlıyorum.
svr_params = {"C": [0.1,0.5,1,3]}
```

```python
# şimdi bir model nesnesi oluştrup buna paremetre değerleri ve hangi modeli bakladiğini ve cv değerlerini ayarlıyorum.
svr_cv_model = GridSearchCV(svr_model, svr_params, cv = 5).fit(X_train, y_train)
```

```python
# Ben modelin çalışma sırasındaki adımları raporlamasını istiyorsam bunu bu şakilde yapmalıyım.
svr_cv_model = GridSearchCV(svr_model, svr_params, cv = 5, verbose=2).fit(X_train, y_train)
```

```python
# peki ben burda bilgisayarımdaki işlemci gücünü maksimumda kulanmak istersem nasıl bir işlem yapmalıyım.
svr_cv_model = GridSearchCV(svr_model, svr_params, cv = 5, verbose=2, n_jobs=-1).fit(X_train, y_train)
```

```python
# Yukardaki model fit etme işleminden sonra ben şimdi burda hangi değerin daha iyi olduğunu burda sorgulamasını yapıyorum.
svr_cv_model.best_params_
```

Final modeli

```python
# Final modeli kuruyorum.
svr_tunned = SVR(kernel = "linear", C = 0.5).fit(X_train, y_train)
```

```python
# Modeli tahmin etirme işlemi gerçekleştiriyorum.
y_pred = svr_tunned.predict(X_test)
```

```python
# Hata kareler ortalamasını hesaplıyorum "RMSE" değeri.
np.sqrt(mean_squared_error(y_test, y_pred))
```
